### This Dataset is about sales of Videos Games across the world in different region with different Genre. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vg = pd.read_csv("vgsales.csv" , index_col="Rank")
df = vg.copy()
```

```python
df.head(90)
```

### Data Cleaning

```python
df.describe()
```

```python
df.columns
```

```python
df.info()
```

```python
#lets first identify if there is any null values or not
df.isnull().sum()
```

```python
# we see that the null values are very less as compared to total data so we can simply remove that null data row from
# the dataset but company pay high cost in getting this data and simply removing it will not be profitable so we can replace
# numeric values with 0.0 and object values with unknown.

df['Year'] = df['Year'].fillna(0.0)
df['Publisher'] = df['Publisher'].fillna("unknown")
```

```python
# Checking if null value exists or not

df.isnull().sum()
```

#### Now our data is cleaned so we will proceed towards visualization and analysis

### Total number of Games made in particular Genre 


```python
df.Genre.value_counts()
```

```python
#representing that on count plot
plt.figure(figsize=(12,10))

sns.countplot(x='Genre'  , data=df )
```

From this graph we can get that Action is most popular game genre which people like the most to play followed by spots and misc genre. For greater sales of the videogame company should build action games as people like it most or companies producing the other genre games should improve their game storyline , graphics ,etc to increase sales.

### Year which has the highest game realease

```python
plt.figure(figsize=(12,10))
sns.countplot(x="Year" , data=df )
plt.xticks(rotation=90)
```

### Year which had the highest sales worldwide

```python
plt.figure(figsize=(12,10))
df_hsw = df.groupby(by=['Year'])['Global_Sales'].sum()
df_hsw = df_hsw.reset_index()
sns.barplot(x="Year" ,y='Global_Sales', data=df_hsw )
plt.xticks(rotation=90)
```

observing the above graph we can se that sales of videogames was increasing and was highest in 2008 and 2009 and thus there were highest number of games releases in these 2 year as company came to know with the highest sales graph that there will be higher demand of videogames in furture.

But we can se that there was huge reduction in both release and sales of video games after year 2009 because after that Online game providing websites and software like STEAM and EpicGames came in popularity and people started buying game online rahter than buying CD's.  

### Game genre which has the highest sale price globally

```python
df_genre = df.groupby(by=['Genre'])['Global_Sales'].sum()
df_genre= df_genre.reset_index()
df_genre= df_genre.sort_values(by=['Global_Sales'], ascending=False)
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x="Genre", y="Global_Sales", data=df_genre)
plt.xticks(rotation=90)
```

We can see the Action genre is mostlly prefered by the audience so the game companies can focus more on Action games for the 
profit and should avoid making Stragtegy genre game which has the lowest sales globaly

### Platfrom have the highest sale price globally

```python
df_genre = df.groupby(by=['Platform'])['Global_Sales'].sum()
df_genre= df_genre.reset_index()
df_genre= df_genre.sort_values(by=['Global_Sales'], ascending=False)
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x="Platform", y="Global_Sales", data=df_genre)
plt.xticks(rotation=90)
```

We can clearly see that PS2 platform is the most preferd platform by the gamers then X360 and the PS3 so the game building companies should make games which support all the three platform PS2 , X360 and PS3 to have maximum sales of their game.
Also the PlayStation companies should work on their device , improve quality and fix bugs so as to maintain the gamers first choise for gaming

### Sales comparison by genre

```python
comp_genre = df[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_map = comp_genre.groupby(by=['Genre']).sum()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(comp_map, annot=True, fmt = '.1f')
```

We can see that NA region loves to video games the most as the NA_sales column has higher values than the others.
Companies are having maximum sales of Sports and Shooter genre of games maximum in the North American region

```python
df1 = comp_map.reset_index()
df1 = pd.melt(df1, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
df1.head()
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x='Genre', y='Sale_Price', hue='Sale_Area', data=df1)
```

Here Action Sports and Shooter have much sales compare with others. and North America (NA_sales) have heights sales all the time. Its look like they love video games.
So game developing comapnies can make those specific genre of games to increase their sales in tha region.


```python
comp_platform = df[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_platform.head()
```

```python
comp_platform = comp_platform.groupby(by=['Platform']).sum().reset_index()
comp_table = pd.melt(comp_platform, id_vars=['Platform'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()
```

```python
plt.figure(figsize=(30, 15))
sns.barplot(x='Platform', y='Sale_Price', hue='Sale_Area', data=comp_table)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

It's clearly shown that X360, Wii, and PS are market leader. and north america always top.

# Statical Analysis

```python
data_hist_log = df.copy()
```

```python
data_hist_log = data_hist_log[data_hist_log.NA_Sales != 0]
data_hist_log = data_hist_log[data_hist_log.EU_Sales != 0]
data_hist_log = data_hist_log[data_hist_log.Other_Sales != 0]
data_hist_log = data_hist_log[data_hist_log.JP_Sales != 0]
data_hist_log = data_hist_log[data_hist_log.Global_Sales != 0]
```

```python
plt.figure(figsize=(25,30))
sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
for i, column in enumerate(sales_columns):
    plt.subplot(3,2,i+1)
    sns.distplot(np.log(data_hist_log[column]), bins=20, kde=True)
```
